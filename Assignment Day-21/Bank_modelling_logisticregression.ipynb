```python
import pandas as pd
import numpy as np
import statsmodels.api as sm
```

```python
bank = pd.read_excel("/home/jatin/ML & AI/Dataset/dataset-20200805T105332Z-001/dataset/Bank_Personal_Loan_Modelling.xlsx",sheet_name=1)
```

```python
bank.info()
```

```python
Y = bank['Personal Loan']
X = bank[['Income','Family','CCAvg','Education','Mortgage','Securities Account','CD Account']]
```

```python
X1 = sm.add_constant(X)
```

```python
Logistic = sm.Logit(Y,X1)
```

```python
result = Logistic.fit()
```

```python
result.summary()
```
