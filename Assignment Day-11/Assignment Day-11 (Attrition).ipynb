```python
stats,p = pearsonr(dataset2.DistanceFromHome,dataset2.Education)
```

```python
stats1,p1 = pearsonr(dataset2.DistanceFromHome,dataset2.PercentSalaryHike)
```

```python
stats3,p3 = pearsonr(dataset2.Age,dataset2.PercentSalaryHike)
```

```python
stats4,p4 = pearsonr(dataset2.Age,dataset2.Education)
```

```python
print(stats4,p4)
```

```python
stats5,p5 = pearsonr(dataset2.MonthlyIncome,dataset2.PercentSalaryHike)
print(stats5,p5)
```
